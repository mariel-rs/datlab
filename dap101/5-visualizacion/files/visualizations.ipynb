# Visualizaciones

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = (10,8)
```

## Matplotlib

```python
# Create an x-axis, and make 2 linear functions of x
import numpy as np

x = np.linspace(-3, 3, 100)
y1 = 3*x
y2 = x**3 + x**2 - x + 1

# Plot both curves on the same axes
fig, ax = plt.subplots()
ax.plot(x, y1, 'g--', label=r'$y_1$')
ax.plot(x, y2, 'y.', label=r'$y_2$')

# .legend() will use the "label" arguments for each curve
ax.legend(loc='lower right', fontsize=14)

plt.show()
```

```python
# Visualización más representativa para mostrar el flujo de visitantes en los meses reportados para un Centro INAH (o estado!) determinado

# Datos Teotihuacan
teotihuacan = inah_df.query('Centro_INAH == "Z. A. Teotihuacan (con Museo de Sitio)"')
teotihuacan
```

```python
meses = ["enero", "febrero", "marzo"]

# Visitantes nacionales
teotihuacan_nal = teotihuacan[["enero_nac", "febrero_nac", "marzo_nac"]]
teotihuacan_nal.columns = meses

# Visitantes internacionales
teotihuacan_ext = teotihuacan[["enero_ext", "febrero_ext", "marzo_ext"]]
teotihuacan_ext.columns = meses

nal = pd.melt(teotihuacan_nal, var_name = "mes", value_name= "visitantes_nacionales")
ext = pd.melt(teotihuacan_ext, var_name = "mes", value_name= "visitantes_extranjeros")
```

```python
nal
```

```python
teotihuacan_df = pd.merge(left = nal, right = ext, how = "inner", on = "mes")
teotihuacan_df
```

```python
sns.lineplot(data = teotihuacan_df, x = "mes", y = "visitantes_nacionales")
```

```python
plot = sns.lineplot(data = teotihuacan_df)

# Modificar las etiquetas de los ejes
plot.set_xlabel("mes", fontsize = 12)
plot.set_ylabel("visitantes", fontsize = 12)

plt.xticks([0, 1, 2], ['enero','febrero','marzo'])

plt.show()
```

```python
x = teotihuacan_df["mes"]
y1 = teotihuacan_df["visitantes_nacionales"]
y2 = teotihuacan_df["visitantes_extranjeros"]

# Plot both curves on the same axes
fig, ax = plt.subplots()
ax.plot(x, y1, 'k-o', label=r'Visitantes nacionales')
ax.plot(x, y2, 'b-o', label=r'Visitantes extranjeros')

ax.set_xlabel(r'Mes', fontsize=12)
ax.set_ylabel(r'Visitantes', fontsize=12)


# .legend() will use the "label" arguments for each curve
ax.legend(loc='upper left', fontsize=12)
plt.show()
```
