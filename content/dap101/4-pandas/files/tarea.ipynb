# Tarea

```python
import pandas as pd
import os
```

```python
# importar datos

data_path = os.path.join(os.getcwd(), "estudiantes_mxuk2021.csv")
estudiantes_df = pd.read_csv(data_path)
```

```python
estudiantes_df.head()
```

## Estudiantes de Puebla que estudian un posgrado

```python
# Una condicion
estudiantes_df[estudiantes_df["estado_origen"] == "PUE"]
```

## Edad promedio de los estudiantes

```python
# Edad promedio de los estudiantes con describe()
estudiantes_df.describe()
```

```python
# Edad promedio de los estudiantes con mean()

estudiantes_df["edad"].mean()
```

## Estudiantes mayores de 27 años y que no son de CDMX

```python
# Multiples condiciones
estudiantes_df[(estudiantes_df["edad"] > 27) & (estudiantes_df["estado_origen"] != "CDMX")]
```

```python
# tambien se puede con query
estudiantes_df.query('edad > 27 & estado_origen != "CDMX"')
```

## Estudiantes menores de 28 años y, que estudian en University of York y en University of Sussex

```python
# Utilizando isin  

universidades = ["University of York", "University of Sussex"]

estudiantes_df[(estudiantes_df["edad"] < 28) & (estudiantes_df["universidad"].isin(universidades))]
```

## Numero de estudiantes por universidad

```python
# Groupby por universidad, contar
top_universidad = estudiantes_df.groupby(by = "universidad").count()
top_universidad
```

## Numero de estudiantes por estado

```python
# Groupby por estado, contar
estudiantes_df.groupby(by = "estado_origen").count()
```

## Universidad con mas estudiantes mexicanos

```python
top_universidad.sort_values(by = "edad", ascending = False)
```

```python
# Un poco mas de detalle
top_universidad.sort_values(by = "edad", ascending = False).head(1)
```
